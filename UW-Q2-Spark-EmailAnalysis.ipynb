```
from pyspark.sql.types import StructType
from pyspark.sql.types import StructField
from pyspark.sql.types import StringType
from pyspark.sql.functions import *

#Set up path variables - modify as needed for multiple sample runs, full-data run, and bonus run
rootdir = 'abfss://pbchamp@uwbigdatatechnologies.dfs.core.windows.net/Week2'

maildir = 'enron_2015_sample' #Sample
#maildir = 'enron_mail_20150507' #Full data

#outputdir = 'enron_2015_sample' #Write Sample to one .csv file
#outputdir = 'enron_2015_sample_output2' #Try writing Sample to multiple .csv files
outputdir = 'enron_2015_sample_output3' #Bonus - parquet
#outputdir = maildir + '_output' #Full data

#Load all files and the path of the file, in case that may be needed (wholeTextFiles loads file path and content)
rdd = sc.wholeTextFiles(rootdir + '/' +  maildir + '/maildir/*/*/*')

#rdd.take(2)

#Option to read data directly into dataframe
#df = spark.read.text(maildir + '/*/*/*', wholetext = True) #wholetext â€“ if true, read each file from input path(s) as a single row.

#Schema for RDD read
schema = StructType([
    StructField("FilePath", StringType(), False),
    StructField("Message", StringType(), False)
])

#Create DataFrame from loaded RDD with schema defined above
df = spark.createDataFrame(rdd, schema)

#Replace my specific path with the generic path
df = df.withColumn('FilePath', regexp_replace('FilePath', rootdir + '/' + maildir + '/', ''))
#df = df.limit(2)

#Fields to parse out of email
fields = ['Message-ID', 'Date', 'From', 'To']

#Regular Expression to get header text of format <FieldName>: <Value>\n (e.g. Message-ID: <18782981.1075855378110.JavaMail.evans@thyme>\n)
regex = "(?<=[field]: )(.*?)(?=\\n)" #using 'Positive Lookbehind', find text between the fieldname colon space and the newline (\n) characters

#Parse and rearrage fields in DataFrame
df = (df
  .withColumn(fields[0], regexp_extract(col('Message'), regex.replace('[field]', fields[0]), 1))
  .withColumn(fields[1], regexp_extract(col('Message'), regex.replace('[field]', fields[1]), 1))
  .withColumn(fields[2], regexp_extract(col('Message'), regex.replace('[field]', fields[2]), 1))
  .withColumn(fields[3], regexp_extract(col('Message'), regex.replace('[field]', fields[3]), 1))
).select(['FilePath', fields[0], fields[1], fields[2], fields[3], 'Message'])
```
