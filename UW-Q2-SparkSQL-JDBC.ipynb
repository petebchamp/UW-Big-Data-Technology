```
#Read parquet data into data frame
path = "abfss://pbchamp@uwbigdatatechnologies.dfs.core.windows.net/Week4/data"
transactionParquetdf = spark.read.parquet(path + "/*.parquet")
```

```
#Review data frame
#transactionParquetdf.rdd.getNumPartitions() #10
transactionParquetdf.count() #2595732 - all records are there
```

```
#Recency: The most recent day from TRANSACTION DATA (can disregard weeks and time)
from pyspark.sql.functions import desc

#display(transactionParquetdf.take(5))
#transactionParquetdf.groupby("household_key").max("DAY").explain()
recency = (transactionParquetdf
  .groupby("household_key")
  .max("DAY")
  .withColumnRenamed("max(DAY)", "Recency")
  .orderBy(desc("Recency")))
display(recency.take(5))
```

```
#Same as above using SQL (just for practice)
transactionParquetdf.createOrReplaceTempView("sales")

recencySQL = spark.sql("""
SELECT household_key, MAX(DAY) AS Recency
FROM sales
GROUP BY household_key
ORDER BY MAX(DAY) DESC""")

display(recencySQL.take(5))
#recencySQL.explain()
```

```
#Frequency: The number of unique basket_id from TRANSACTION_DATA
frequency = ((transactionParquetdf
   .select("household_key", "BASKET_ID")
   .distinct())
      .groupBy("household_key")
      .count()
      .withColumnRenamed("count", "Frequency")
      .orderBy(desc("Frequency")))
display(frequency.take(5))
```

```
#Same as above using SQL (just for practice)
frequencySQL = spark.sql("""
SELECT household_key, COUNT(DISTINCT(BASKET_ID)) Frequency
FROM sales
GROUP BY household_key
ORDER BY Frequency DESC
""")
display(frequencySQL.take(5))
```

```
#Monetary Value: The sum of all sales_value from TRANSACTION_DATA
from pyspark.sql.functions import round

monetary = (transactionParquetdf
  .groupBy("household_key")
  .sum("SALES_VALUE")
  .withColumnRenamed("sum(SALES_VALUE)", "Monetary")
  .withColumn("Monetary", round("Monetary", 2))
  .orderBy(desc("Monetary")))

display(monetary.take(5))
```

```
#Same as above using SQL (just for practice)
monetarySQL = spark.sql("""
SELECT household_key, ROUND(SUM(SALES_VALUE), 2) Monetary
FROM sales
GROUP BY household_key
ORDER BY Monetary DESC
""")
display(monetarySQL.take(5))
```

```
#recency.count() #2500
#frequency.count() #2500
#monetary.count() #2500

#One way to join if key columns are named diffent
#rfm = recency.join(frequency, recency.household_key == frequency.household_key) #duplicates household_key

#Join three tables and sort
rfm = recency.join(frequency.join(monetary, "household_key"), "household_key").orderBy(desc("Recency"), desc("Frequency"), desc("Monetary"))
#rfm.count() #2500
display(rfm.take(5))
```

```
#Bonus Exercise #2

#Calculate quantile values (with approxQuantile) for recency, frequency, and monetary and load into a dataframe
#Include a rownumber (to be used later in a join)

from pyspark.sql.functions import row_number,lit
from pyspark.sql.window import Window

columns = ["Recency", "Frequency", "Monetary"]
percentiles = [0.0, .33, .66, 1.0]
w = Window().orderBy(lit("Percentile")) #Could be anything, apparently

quantile_benchmarks = (spark.createDataFrame(
    zip(percentiles, *rfm.approxQuantile(columns, percentiles, 0.0)), 
    ["Percentile"] + columns
)).withColumn("ROWNUM", row_number().over(w))

display(quantile_benchmarks)
```

```
#Get min and max quantile ranges
#Use self join on ROWNUM, although could use lead window function (see below example)

quantile_benchmarks.createOrReplaceTempView("quantile_benchmarks")

quantile_benchmark_ranges = spark.sql("""
SELECT 
  q1.ROWNUM - 1 Quantile, 
  q1.Percentile, 
  q1.Recency MinRecency, 
  q2.Recency - 1 MaxRecency,
  q1.Frequency MinFrequency, 
  q2.Frequency - 1 MaxFrequency,
  q1.Monetary MinMonetary, 
  q2.Monetary - .01 MaxMonetary
FROM quantile_benchmarks q1
INNER JOIN quantile_benchmarks q2
  ON q1.ROWNUM + 1 = q2.ROWNUM
""")

display(quantile_benchmark_ranges)
```

```
#EXAMPLE lead function 
#Could be enahnced to replace above SQL self join

from pyspark.sql.window import Window
from pyspark.sql.functions import lead

test = (quantile_benchmarks
  .withColumnRenamed('Recency', 'RecencyMin')
  .withColumn('RecencyMax', lead('RecencyMin')              
    .over(Window.orderBy('Percentile')
    )
))

display(test)
```

```
#Calculate quantile scores (0, 1, 2)
#Need left outer join because the quantile 2 max is just less than the actual maxes
# and default nulls to #2.

rfm.createOrReplaceTempView("rfm")
quantile_benchmark_ranges.createOrReplaceTempView("quantile_benchmark_ranges")

rfm_quantiles = spark.sql("""
SELECT
  rfm.*, 
  COALESCE(qr.Quantile, 2) r_score, 
  COALESCE(qf.Quantile, 2) f_score,
  COALESCE(qm.Quantile, 2) m_score
FROM rfm 
LEFT OUTER JOIN quantile_benchmark_ranges qr
  ON rfm.Recency BETWEEN qr.MinRecency AND qr.MaxRecency
LEFT OUTER JOIN quantile_benchmark_ranges qf
  ON rfm.Frequency BETWEEN qf.MinFrequency AND qf.MaxFrequency
LEFT OUTER JOIN quantile_benchmark_ranges qm
  ON rfm.Monetary BETWEEN qm.MinMonetary AND qm.MaxMonetary   
""")

display(rfm_quantiles.take(5))
```
